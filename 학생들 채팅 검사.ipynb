```python
chat = []
result = []

with open (directory + '\\' + today_dir[-1] + '\\'+ 'meeting_saved_chat.txt', encoding='UTF8') as f:
    chat = f.read().split('\n')

for i in range(0,len(chat),2):
    try :
        result.append([chat[i], chat[i+1].replace('\t','')])
    except :
        pass
print(result)
```
